# Generate Audio

```python
pipe = StableAudioPipeline.from_pretrained("stabilityai/stable-audio-open-1.0", torch_dtype=torch.float16)
pipe = pipe.to("cuda")

# define the prompts
prompt = "Generate a soundscape of a jungle, rich with nature sounds. The atmosphere should be dense with the rustling of trees, leaves, and branches swaying. Camouflaged animals blend into the environment, through their sounds—soft bird calls, distant owl hoots, occasional gecko chirps, the rhythmic croaking of frogs, and the delicate flapping of butterfly wings"
negative_prompt = "Low quality."

# set the seed for generator
generator = torch.Generator("cuda")

# run the generation, requesting 4 waveforms
audio = pipe(
    prompt,
    negative_prompt=negative_prompt,
    num_inference_steps=200,
    audio_end_in_s=21.0,
    num_waveforms_per_prompt=4,  # Request 4 waveforms
    generator=generator,
).audios

# Iterate through generated audio and save each to a separate file
for i, waveform in enumerate(audio):
    output = waveform.T.float().cpu().numpy()
    sf.write(f"sound_{i}.wav", output, pipe.vae.sampling_rate)

    # Optionally, play each audio after saving (for Colab/Jupyter)
    # display(Audio(f"output_{i}.wav"))
```
